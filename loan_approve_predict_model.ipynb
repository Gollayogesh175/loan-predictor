```python
import pandas as pd
```

```python
data = pd.read_csv('loan_approve_dataset.csv')
```

```python
data
```

```python
data.columns
```

```python
data.isnull().sum
```

```python
from sklearn.model_selection import train_test_split
```

```python
input_data = data.drop(columns=['Loan_Status'])
output_data = data['Loan_Status']
```

```python
input_data
```

```python
output_data
```

```python
x_train,x_test,y_train,y_test = train_test_split(input_data,output_data, test_size=0.2)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
```

```python
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train_scaled, y_train)
```

```python
model.score(x_test_scaled, y_test)
```

```python
pred_data = pd.DataFrame([['1','1','1','1','0','550000','150000','360','750','1000000']],columns=['Gender','Marital_Status','Dependents','Education','Self_Employed','Applicant_Income','Loan_Amount','Loan_Term','Cibil_score','Assets'])
```

```python
pred_data = scaler.transform(pred_data)
```

```python
model.predict(pred_data)
```
